## Task 1: User Login Check

```python
def user_login_check():
    correct_username = "admin"
    correct_password = "1234"

    username = input("Enter Username: ")
    password = input("Enter Password: ")

    if username == correct_username and password == correct_password:
        print("Login Successful")
    else:
        print("Invalid Credentials")

user_login_check()
```

## Task 2: Pass / Fail Analyzer



```python
def pass_fail_analyzer():
    marks_input = input("Enter marks separated by spaces: ")
    marks = list(map(int, marks_input.split()))

    pass_count = 0
    fail_count = 0

    for mark in marks:
        if mark >= 50:
            pass_count += 1
        else:
            fail_count += 1

    print("\nResult Summary:")
    print("Total Students:", len(marks))
    print("Passed Students:", pass_count)
    print("Failed Students:", fail_count)

pass_fail_analyzer()
```

## Task 3: Simple Data Cleaner

```python
def simple_data_cleaner():
    names_input = input("Enter names separated by commas: ")
    names = names_input.split(",")

    cleaned_names = []

    for name in names:
        cleaned_names.append(name.strip().lower())

    print("Cleaned Names:")
    print(cleaned_names)

simple_data_cleaner()
```

## Task 4: Message Length Analyzer

```python
def message_length_analyzer():
    messages_input = input("Enter messages separated by commas: ")
    messages = messages_input.split(",")

    print("\nMessage Analysis:")

    for message in messages:
        clean_message = message.strip()
        length = len(clean_message)

        print(f"Message: '{clean_message}' | Length: {length}")

        if length > 10:
            print("⚠️ This message is longer than 10 characters.")

message_length_analyzer()
```

## Task 5: Error Message Detector


```python
def error_message_detector():
    logs_input = input("Enter log entries separated by commas: ")
    logs = logs_input.split(",")

    error_count = 0

    for log in logs:
        if log.strip().upper() == "ERROR":
            error_count += 1

    print("Total ERROR entries:", error_count)

error_message_detector()
```
