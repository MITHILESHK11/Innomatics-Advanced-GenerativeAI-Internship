## Problem 1: Mobile Recharge Validation System

```python
def recharge_validation():
    valid_plans = [199, 299, 399, 599]
    while True:
        amount = int(input("Enter recharge amount: "))
        if amount < 50:
            print("Recharge amount must be at least ₹50")
        elif amount not in valid_plans:
            print("Invalid plan selected")
        else:
            print("Recharge Successful")
            break
```

## Problem 2: Inventory Reorder Alert System

```python
def inventory_check(products):
    for product, stock in products.items():
        if stock < 15:
            print(product, "→ Reorder Alert")
        else:
            print(product, "→ Stock OK")
```

## Problem 3: Student Result Processing System

```python
def student_result(marks):
    total = 0
    for mark in marks:
        total += mark
    average = total / len(marks)
    if average >= 50:
        return "Pass"
    else:
        return "Fail"
```

## Problem 4: Cab Fare Estimator with Retry Option

```python
def calculate_fare(distance, peak):
    fare = 50 + (12 * distance)
    if peak:
        fare = fare + (fare * 0.25)
    return fare

def cab_fare_system():
    while True:
        distance = float(input("Enter distance in km: "))
        peak_input = input("Is it peak hour? (yes/no): ").lower()
        peak = True if peak_input == "yes" else False
        fare = calculate_fare(distance, peak)
        print("Total Fare:", fare)
        retry = input("Do you want to retry? (yes/no): ").lower()
        if retry != "yes":
            break
```

## Problem 5: Employee Attendance Eligibility Checker


```python
def attendance_check(attendance):
    present = 0
    for day in attendance:
        if day == "P":
            present += 1
    percentage = (present / len(attendance)) * 100
    if percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"
```

## Problem 6: Password Strength Checker

```python
def password_checker(password):
    special_chars = "@#$"
    has_digit = False
    has_special = False

    if len(password) < 8:
        return "Weak Password"

    for char in password:
        if char.isdigit():
            has_digit = True
        if char in special_chars:
            has_special = True

    if has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak Password"
```
