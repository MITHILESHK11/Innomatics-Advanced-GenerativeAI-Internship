## Problem 1: Social Media – Post Engagement Analyzer

```python
def post_engagement(likes):
    total_likes = 0
    for like in likes:
        total_likes += like
    print("Total Likes:", total_likes)
    if total_likes >= 1000:
        print("Post Status: Viral Post")
    else:
        print("Post Status: Normal Engagement")

post_engagement([400, 350, 300])
```

## Problem 3: Agriculture – Rainfall Adequacy Checker

```python
def rainfall_check(rainfall_data, required_level):
    total = 0
    for value in rainfall_data:
        total += value
    average = total / len(rainfall_data)
    print("Average Rainfall:", average)
    if average >= required_level:
        print("Rainfall Status: Adequate Rainfall")
    else:
        print("Rainfall Status: Inadequate Rainfall")

rainfall_check([60, 75, 80], 70)
```

## Problem 4: Social Media – Duplicate Account Detection

```python
def duplicate_accounts(usernames):
    unique_users = set(usernames)
    if len(unique_users) < len(usernames):
        print("Duplicate Accounts Found: Yes")
    else:
        print("Duplicate Accounts Found: No")

duplicate_accounts(["user1", "user2", "user1"])
```

## Problem 9: E-Learning – Student Attendance Eligibility System

```python
def attendance_eligibility(attendance):
    present = 0
    for day in attendance:
        if day == "P":
            present += 1
    percentage = (present / len(attendance)) * 100
    print("Attendance Percentage:", percentage)
    if percentage >= 75:
        print("Exam Eligibility: Eligible")
    else:
        print("Exam Eligibility: Not Eligible")

attendance_eligibility(["P","P","A","P","P"])
```

## Problem 10: Smart Electricity Bill Analyzer

```python
def electricity_bill(units):
    bill = 0
    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + ((units - 100) * 5)
    else:
        bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)

    if bill < 500:
        status = "Low Usage"
    elif bill <= 1500:
        status = "Moderate Usage"
    else:
        status = "High Usage"

    return bill, status

bill, status = electricity_bill(250)
print("Total Bill:", bill)
print("Usage Level:", status)
```
